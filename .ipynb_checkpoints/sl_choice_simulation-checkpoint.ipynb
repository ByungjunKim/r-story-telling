## 2. 황동석 편 - ep1_dongseok

```python
import pandas as pd

tsv_path_updated = "ep1_dongseok.tsv"
df_do = pd.read_csv(tsv_path_updated, sep="\t")

# 데이터 확인
df_do.head()
```

```python
import json

jsonl_output_path = "ep1_dongseok_prompt.jsonl"

first_row = df_do[df_do["사건명"].notna()].iloc[0]
title = "황동석 편"
situation = str(first_row["상황"]).strip()

skills = [str(skill).strip() for skill in df_do["스킬"] if pd.notna(skill)]
extra = [str(desc).strip() for desc in df_do["부가 설명"] if pd.notna(desc)]

dialogue = [str(line).strip() for line in df_do["대사"] if pd.notna(line)]

choices = [
    "1. 적을 모두 처치한다",
    "2. 무력화하거나 기절시킨다",
    "3. 도망친다"
]
question = "성진우는 어떤 선택을 할까? 그리고 그 이유는 무엇일까?"

json_obj = {
    "title": title,
    "situation": situation,
    "skills": skills,
    "extra": extra,
    "persona_hint": "",  #  추가 가능
    "dialogue": dialogue,
    "choices": choices,
    "question": question
}

with open(jsonl_output_path, "w", encoding="utf-8") as f:
    json.dump(json_obj, f, ensure_ascii=False)
    f.write("\n")

jsonl_output_path


```

```python
def format_prompt(entry):
    return f"""상황: {entry['situation']}
스킬: {entry['skills']}
부가 설명: {entry['extra']}
성진우의 대사:
{chr(10).join(entry['dialogue'])}

선택지:
{chr(10).join(entry['choices'])}

{entry['question']}
"""

#프롬프트 생성함수
```

```python
import json

with open("ep1_dongseok_prompt.jsonl", "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

entry = data[0]

print(format_prompt(entry))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import json

model_name = "kakaocorp/kanana-nano-2.1b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).cuda() 

jsonl_path = "ep1_dongseok_prompt.jsonl" #황동석 사건 모음 
with open(jsonl_path, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]


def format_prompt(entry):
    return f"""상황: {entry['situation']}
스킬: {entry['skills']}
부가 설명: {entry['extra']}
성진우의 대사:
{chr(10).join(entry['dialogue'])}

선택지:
{chr(10).join(entry['choices'])}

{entry['question']}
"""

entry = data[0]  # 황동석 사건 하나
prompt = format_prompt(entry)

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=1028,
        temperature=0.8,
        top_p=0.9,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

response = tokenizer.decode(outputs[0], skip_special_tokens=True)[len(prompt):].strip()

print(response)

output_path = "prompt_output/ep1_dongseok_result.md"
with open(output_path, "w", encoding="utf-8") as f:
    f.write(response)
```

## 3. 히든 퀘스트 편 - ep2_hidden_quest

```python
def format_prompt(entry):
    return f"""상황: {entry['situation']}
스킬: {entry['skills']}
부가 설명: {entry['extra']}
성진우의 대사:
{chr(10).join(entry['dialogue'])}

선택지:
{chr(10).join(entry['choices'])}

{entry['question']}
"""
```

```python
import pandas as pd
import json

tsv_path = "ep2_hidden.tsv"
jsonl_output_path = "ep2_hidden_prompt.jsonl"

df_hidden = pd.read_csv(tsv_path, sep='\t')

first_row = df_hidden[df_hidden["사건명"].notna()].iloc[0]

title = str(first_row["사건명"]).strip() if pd.notna(first_row["사건명"]) else "히든 퀘스트 편"
situation = str(first_row["상황"]).strip() if pd.notna(first_row["상황"]) else ""

skills = [str(skill).strip() for skill in df_hidden["스킬"] if pd.notna(skill)]
extra = [str(desc).strip() for desc in df_hidden["부가 설명"] if pd.notna(desc)]

dialogue = [str(line).strip() for line in df_hidden["대사"] if pd.notna(line)]

choices = [
    "1. 축복 받은 랜덤 박스를 선택한다.",
    "2. 저주 받은 랜덤 박스를 선택한다."
]
question = "성진우는 어떤 박스를 선택할까? 그리고 그 이유는?"

json_obj = {
    "title": title,
    "situation": situation,
    "skills": skills,
    "extra": extra,
    "persona_hint": "",  
    "dialogue": dialogue,
    "choices": choices,
    "question": question
}

# 저장
with open(jsonl_output_path, "w", encoding="utf-8") as f:
    json.dump(json_obj, f, ensure_ascii=False)
    f.write("\n")

print(f"저장 : {jsonl_output_path}")
```

```python
def format_prompt(entry):
    return f"""상황: {entry['situation']}
스킬: {'; '.join(entry['skills'])}
부가 설명: {'; '.join(entry['extra'])}
성진우의 대사:
{chr(10).join(entry['dialogue'])}

선택지:
{chr(10).join(entry['choices'])}

{entry['question']}
"""
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "kakaocorp/kanana-nano-2.1b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to("cuda" if torch.cuda.is_available() else "cpu")

with open("ep2_hidden_prompt.jsonl", "r", encoding="utf-8") as f:
    entry = json.loads(f.readline())

prompt = format_prompt(entry)

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=1028,
        do_sample=True,
        top_p=0.95,
        temperature=0.8,
        pad_token_id=tokenizer.eos_token_id
    )

response = tokenizer.decode(outputs[0], skip_special_tokens=True)[len(prompt):].strip()
print(response)

output_path = "prompt_output/ep2_hidden_result.md"
with open(output_path, "w", encoding="utf-8") as f:
    f.write(response)
```

## 4.전직 퀘스트 - ep3_job_quest


```python
import pandas as pd
import json
import os

df_job = pd.read_csv("ep3_job_quest.tsv", sep="\t")

first_row = df_job[df_job["사건명"].notna()].iloc[0]

title = "전직 퀘스트 편"
situation = str(first_row["상황"]).strip()
skills = [str(skill).strip() for skill in df_job["스킬"] if pd.notna(skill)]
extra = [str(e).strip() for e in df_job["부가 설명"] if pd.notna(e)]
dialogue = [str(line).strip() for line in df_job["대사"] if pd.notna(line)]

choices = [
    "1. 히든 클래스 '네크로맨서'로 전직한다.",
    "2. 기본 클래스로 전직하고 던전을 빠르게 탈출한다."
]

question = "성진우는 어떤 선택을 할까? 그리고 그 이유는 무엇일까?"

json_obj = {
    "title": title,
    "situation": situation,
    "skills": skills,
    "extra": extra,
    "persona_hint": "",
    "dialogue": dialogue,
    "choices": choices,
    "question": question
}

jsonl_path = "ep3_job_quest_prompt.jsonl"

with open(jsonl_path, "w", encoding="utf-8") as f:
    json.dump(json_obj, f, ensure_ascii=False)
    f.write("\n")

print(f"저장: {jsonl_path}")
```

```python
def format_prompt(entry):
    return f"""상황: {entry['situation']}
스킬: {'; '.join(entry['skills'])}
부가 설명: {'; '.join(entry['extra'])}
성진우의 대사:
{chr(10).join(entry['dialogue'])}

선택지:
{chr(10).join(entry['choices'])}

{entry['question']}
"""
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "kakaocorp/kanana-nano-2.1b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to("cuda" if torch.cuda.is_available() else "cpu")
# 프롬프트 로드
with open("ep3_job_quest_prompt.jsonl", "r", encoding="utf-8") as f:
    entry = json.loads(f.readline())

prompt = format_prompt(entry)

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=1024,
        do_sample=True,
        top_p=0.95,
        temperature=0.8,
        repetition_penalty=1.1, #반복억제
        pad_token_id=tokenizer.eos_token_id
    )

response = tokenizer.decode(outputs[0], skip_special_tokens=True)[len(prompt):].strip()
print(response)

output_path = "prompt_output/ep3_job_quest_result.md"
with open(output_path, "w", encoding="utf-8") as f:
    f.write(response)
```
